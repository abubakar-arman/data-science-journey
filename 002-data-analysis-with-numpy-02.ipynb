### Manipulating ndarray

```python
import numpy as np

## Reshaping

arr = np.arange(20).reshape(4,5)
print(arr)
print(arr.shape)

arr = arr.reshape(2, 10)
print(arr)
print(arr.shape)
```

```python
## Flattening
print(arr.ndim)

arr = arr.flatten()
print(arr)
print(arr.shape)
print(arr.ndim)
```

```python
## Raveling
arr = np.arange(20).reshape(4,5)
print(arr)
print(arr.ndim)

flatten_row_wise = arr.ravel()
print(flatten_row_wise)

flatten_col_wise = arr.ravel(order='F')
print(flatten_col_wise)
```
