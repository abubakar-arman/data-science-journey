### Manipulating ndarray

```python
import numpy as np

## Reshaping

arr = np.arange(20).reshape(4,5)
print(arr)
print(arr.shape)

arr = arr.reshape(2, 10)
print(arr)
print(arr.shape)
```

```python
## Flattening
print(arr.ndim)

arr = arr.flatten()
print(arr)
print(arr.shape)
print(arr.ndim)
```

```python
## Raveling
arr = np.arange(20).reshape(4,5)
print(arr)
print(arr.ndim)

flatten_row_wise = arr.ravel()
print(flatten_row_wise)

flatten_col_wise = arr.ravel(order='F')
print(flatten_col_wise)
```

```python
## Concatenation

a1 = np.arange(20).reshape(4,5)
a2 = np.arange(20, 40).reshape(4,5)
print(a1, a2, sep='\n')

concat_row_wise = np.concatenate((a1, a2))
concat_col_wise = np.concatenate((a1, a2), axis=1)
print()
print(concat_row_wise)
print(concat_col_wise)
```

```python
## Transpose

a = np.arange(20).reshape(2,10)
print(a)
print(a.T)
```

```python
## Array Splitting

a = np.arange(20)
print(a)

a_splitted = np.array_split(a, 3)
print(a_splitted)

a_equal_splitted = np.split(a, 4)
print(a_equal_splitted)
```

### ndarray Arithmetic Operators

```python
a1 = [10, 20, 30, 40, 50]
a2 = [2, 3, 4, 5, 6]

# faster for vectorization
b1 = np.array(a1)
b2 = np.array(a2)

## Operators

print( a1+a1 ) # list concat
print( b1+b2 ) # element wise addition ## np.add(b1, b2)
print()
# print( a1*a2 ) # error multiplying by a list
print( b1*b2 ) # element wise multiplication

print( b1%b2 )
```

### Trigonometric Functions

```python

print(b2)
print(np.sin(b2))
print(np.cos(b2))
print(np.tan(b2))
print(np.rad2deg(b2))


print()
print(b1)
print(np.log10(b1)) # it means -> oi element ke 10 diye koybar vag kora jabe
print(np.log2(b1)) # 2 diye koybar vag kora jabe -> 40 - 20 - 10 - 5 - 2.5
print(np.sqrt(b1)) 
```

```python
print(b2)
print(np.sum(b2))
print(np.cumulative_sum(b2))
```

### Logical Operator

```python
print(b1)
print(b2)

greater = b1 > b2
print(greater)

equal = b1==b2
print(equal)

print(np.all(greater))
print(np.any(equal))
```

### Sorting

```python
# 1D sort
x = np.random.choice(10, 10, replace=False)
print(x)

y = x.copy()
print(y)

y.sort() # inplace sort
print(y)
print(y)
print()

z = np.sort(x) # copy sort
print(z)
print(x)
```

```python
# 2D sort
a = np.random.choice(16, (4,4), replace=False)
print(a)

a_sorted_row_wise = np.sort(a) # row-wise sort (axis=1)
print(a_sorted_row_wise)

a_sorted_col_wise = np.sort(a, axis=0)
print(a_sorted_col_wise)
```

### Searching 

```python
a = np.array([
    [1, 9, 3],
    [4, 5, 6],
    [7, 8, 9]
])
print(a)

print(np.where(a==9)) # returns indices
print(np.where(a<6))

print(np.where(a < 6, a, 0)) # returns array
```

```python
# find out max value index
b = np.random.choice(10, 10)
print(b)
print(np.argmax(b)) # returns max val index
print(np.argmin(b))
```

### Counting

```python
a = np.random.choice(100, (100))
print(a)

print(np.count_nonzero(a>50))
print(np.count_nonzero(a=50))
```

```python
print(np.unique(a))
unique_val, counts = np.unique(a, return_counts=True)
print(unique_val)
print(counts)
```

### Statistical Functions

```python
marks = np.genfromtxt('dataset/student_scores.csv', delimiter=',', skip_header=True) # load data
print(marks)
```

```python
marks_math = marks[:, :1]
marks_sci = marks[:, 1:2]
marks_eng = marks[:, 2:]

print(marks_math.T) # transpose it for better view
```

```python
print(np.max(marks_math))
print(np.min(marks_math))
print(np.average(marks_math))
print(np.mean(marks_math))
print(np.median(marks_math))
print(np.std(marks_math))
```

```python
## Correlation
study_hours = np.array([2,4,6,8,9])
marks = np.array([49,55,60,90,99])

table = np.array([study_hours, marks])
print(table)
print(np.corrcoef(table))
```

### Linear Algebra with Numpy

```python
a = np.arange(6).reshape(2,3)
print(a)
b = np.arange(6).reshape(3,2)
print(b)
```

```python
print(np.dot(a,b))
print(np.trace(a))

mat_square = np.random.choice(9, (3,3))
print(mat_square)
print(np.linalg.det(mat_square))
print(np.linalg.matrix_rank(mat_square))
```
