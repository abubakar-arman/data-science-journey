### Manipulating ndarray

```python
import numpy as np

## Reshaping

arr = np.arange(20).reshape(4,5)
print(arr)
print(arr.shape)

arr = arr.reshape(2, 10)
print(arr)
print(arr.shape)
```

```python
## Flattening
print(arr.ndim)

arr = arr.flatten()
print(arr)
print(arr.shape)
print(arr.ndim)
```

```python
## Raveling
arr = np.arange(20).reshape(4,5)
print(arr)
print(arr.ndim)

flatten_row_wise = arr.ravel()
print(flatten_row_wise)

flatten_col_wise = arr.ravel(order='F')
print(flatten_col_wise)
```

```python
## Concatenation

a1 = np.arange(20).reshape(4,5)
a2 = np.arange(20, 40).reshape(4,5)
print(a1, a2, sep='\n')

concat_row_wise = np.concatenate((a1, a2))
concat_col_wise = np.concatenate((a1, a2), axis=1)
print()
print(concat_row_wise)
print(concat_col_wise)
```

```python
## Transpose

a = np.arange(20).reshape(2,10)
print(a)
print(a.T)
```

```python
## Array Splitting

a = np.arange(20)
print(a)

a_splitted = np.array_split(a, 3)
print(a_splitted)

a_equal_splitted = np.split(a, 4)
print(a_equal_splitted)
```
