```python
import numpy as np

arr = np.array([1, 2, 3, 4])

print(arr, type(arr))

print(arr * 2) # ndarray supports whole array computation
```

### Dimensions of ndarray

```python
arr1d = np.array([1, 2, 3]) # single row or vector
arr2d = np.array([ # array inside array | a sheet of paper
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
])
arr3d = np.array([ # whole building
    [ # 1st floor of the building
        [1, 2, 3], # 1st apartment on 1st floor
        [4, 5, 6]
    ], 
    [ # 2nd floor 
        [7, 8, 9], # 1st aptment on 2nd floor
        [10, 11, 12]
    ]
])
```

### Attributes of ndarray

```python
print(arr1d.ndim, arr2d.ndim, arr3d.ndim) # .ndim
print(arr1d.shape, arr2d.shape, arr3d.shape) # .shape
print(arr1d.size, arr2d.size, arr3d.size) # .size | total element count
```

### Datatypes of   ndarray 

```python
arr = np.array([1, 2, 3])
print(arr.dtype)

arr = np.array([1.1, 2.2, 3.3])
print(arr.dtype)

arr = np.array(['aaaaa', 'bbb'])
print(arr.dtype)
```

```python
arr = np.array([1, 2, 3.3]) # upcasted to float
print(arr.dtype)

arr = np.array([1, 1.4, False, 2, '3']) # upcasted to string
print(arr.dtype)
```

### Selecting datatype for an array

```python
arr = np.array([1, 2, 30])
print(arr.dtype)

arr = np.array([1, 3, 5], dtype=np.int8)
print(arr.dtype)

# arr = np.array([1, 3, 128], dtype=np.int8) # int8 supports -128 to 127 only
# print(arr.dtype)

arr = arr.astype(np.float64) # ndarray.astype()
print(arr)
print(type(arr))
print(arr.dtype)
```

### Creating ndarray from existing data

```python
lst = [1, 2, 3]

arr = np.array(lst)
print(arr)
print(arr.dtype)

arr = np.array(lst, dtype=np.float64)
print(arr.dtype)

mixed_lst = [1, 3.3, 'asdf'] # Upcasting happens
arr = np.array(mixed_lst)
print(arr.dtype)
```

```python
matrix = [
    [1,2,3],
    [4,5,6]
]

arr = np.array(matrix)

print(arr)
print(arr.dtype)
```

```python
tuple = (1, 2,3)

print(tuple)
print(type(tuple))

arr = np.array(tuple, dtype=np.float32)
print(type(arr))
```

```python
set = {1,2,3} # set is abnormal, rather convert to list
print(set)
print(type(set))

arr = np.array(set)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.ndim)
print(arr.shape)
print(arr.size)
```

```python
arr = np.array(list(set))

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.ndim)
print(arr.shape)
print(arr.size)
```

```python
# from dictionary
dct = {'a': 1, 'b': 2, 'c': 3}

keys = dct.keys()
values = dct.values()
items = dct.items()

print(keys)
print(values)
print(items)

arr = np.array(dct) # direct dict input is wrong approach

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.ndim)
print(arr.shape)
print(arr.size)
```

```python
arr = np.array(keys)
print('\n', arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.size)

arr = np.array(values)
print('\n', arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.size)

arr = np.array(items)
print('\n', arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python

arr = np.array(list(keys)) # initialize after converting to list
print('\n', arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.size)

arr = np.array(list(values))
print('\n', arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.size)

arr = np.array(list(items))
print('\n', arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

### ndarray generation with random value

```python
arr = np.random.rand(2,3) # outputs random float b/w 0-1
print(arr)
print(arr.dtype)

arr = np.random.randint(10, 20, (2, 3)) # integer
print(arr)
print(arr.dtype)

np.random.uniform(10, 20, (3,4)) # floating point
```
