### Dimensions of ndarray

```python
arr1d = np.array([1, 2, 3]) # single row or vector
arr2d = np.array([ # array inside array | a sheet of paper
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
])
arr3d = np.array([ # whole building
    [ # 1st floor of the building
        [1, 2, 3], # 1st apartment on 1st floor
        [4, 5, 6]
    ], 
    [ # 2nd floor 
        [7, 8, 9], # 1st aptment on 2nd floor
        [10, 11, 12]
    ]
])
```

### Attributes of ndarray

```python
print(arr1d.ndim, arr2d.ndim, arr3d.ndim) # .ndim
print(arr1d.shape, arr2d.shape, arr3d.shape) # .shape
print(arr1d.size, arr2d.size, arr3d.size) # .size | total element count
```

### Datatypes of   ndarray 

```python
arr = np.array([1, 2, 3])
print(arr.dtype)

arr = np.array([1.1, 2.2, 3.3])
print(arr.dtype)

arr = np.array(['aaaaa', 'bbb'])
print(arr.dtype)
```

### Selecting datatype for an array

```python
arr = np.array([1, 2, 30])
print(arr.dtype)

arr = np.array([1, 3, 5], dtype=np.int8)
print(arr.dtype)

# arr = np.array([1, 3, 128], dtype=np.int8) # int8 supports -128 to 127 only
# print(arr.dtype)

arr = arr.astype(np.float64) # ndarray.astype()
print(arr)
print(type(arr))
print(arr.dtype)
```

### Indexing and Slicing

```python
### Indexing

arr1 = np.array([1, 2, 3])
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(arr[1])
print(arr2[0])
```

```python
### Slicing
# 1D array
arr = np.random.randint(1, 100, (10))
print(arr)

arr_view = arr[2:8] # creates a view, not copy
print(arr_view)

arr_view[0] = 10000
print(arr)

arr_copy = arr[2:8].copy() # makes a copy, not reference
print(arr_copy)

arr_copy[0] = -2000
print(arr_copy)
print(arr)
```

```python
# 2D array
arr = np.arange(25).reshape(5,5)
print(arr)

print(arr[:3]) # select first 3 rows
print(arr[:, 3:]) # select last 2 cols
print(arr[1:4, 1:4]) # select submatrix from middle
```

### Advanced Indexing

```python
arr = np.arange(10, 20)
print(arr)
# 1D
print(arr[[3,4,8]])

arr = np.arange(9).reshape(3,3)
print(arr)
# 2D
print(arr[[1,1,2], [1,2,2]]) # arr[[all rows], [all cols]] 
# using bools
print(arr[arr>4])

arr[arr>3] = 0
print(arr)
```

### Iteration

### Creating dataset

```python
# given dataset
data = np.arange(30).reshape(10,3)
print(data)

# randomizing dataset
random_indices = np.random.choice(10, (10), replace=False)
print(random_indices)

random_dataset = data[random_indices]
print(random_dataset)
```

```python
# Extraction of Features and Target

X = random_dataset[:, :2] # features
print(X)

y = random_dataset[:, 2:] # target
print(y)
```

```python
# Train Test Splitting
train_index = int(len(data) * 0.8)
print(train_index)

print(X)

X_train = X[ : train_index]
print(X_train)

X_test = X[train_index : ]
print(X_test)

y_train = y[:train_index]
y_test = y[train_index:]
print(y_train)
print(y_test)
```
