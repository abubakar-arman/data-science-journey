## Univariate analysis for categorical columns

### Analyze survived column

```python
sns.countplot(data=df, x=df['Survived'])
```

```python
df['Survived'].value_counts() / len(df) *100 # get percentage
```

```python
survived_count = df["Survived"].value_counts()
print(survived_count)
```

```python
survived_label = df['Survived'].unique()
survived_label
```

```python
type(survived_count)
```

```python
plt.pie(survived_count, autopct='%1.1f', labels=survived_label)
plt.title("Pi chart of survival rate")
```

## Univariate analysis for Numerical Columns

### Univariate analysis for Age column

```python
sns.histplot(df['Age'])

plt.title('Histogram of Age')
plt.xlabel('Age')
plt.show()
```

```python

sns.histplot(df['Age'], bins=50)

plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
# plt.show()
```

```python
sns.kdeplot(df['Age'])

plt.title('KDE plot for Age')
plt.xlabel('Ages')
plt.ylabel('Propability Distribution / Density')
```

```python
sns.boxplot(df['Age'])
```

### Univariate analysis for Fare

```python
sns.histplot(df['Fare'], bins=50)

plt.title('Histplot for Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
sns.kdeplot(df['Fare'])

plt.title('KDE plot for Fare')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
```

```python
sns.boxplot(df['Fare'])
```
